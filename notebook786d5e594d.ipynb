```python
from sklearn.metrics import jaccard_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split as tts
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/kaggle/input/diabetes-dataset-for-beginners/diabetes.csv')
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

print(X)
print(y)

print(data.isna().values.any())

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.15, random_state=2)

LR = LogisticRegression(C=0.001, solver="liblinear").fit(X_train, y_train)

print(LR)
yhat = LR.predict(X_test)
print(yhat)

print(jaccard_score(y_test, yhat, pos_label=0))

from sklearn.neighbors import KNeighborsClassifier as knn
classifier=knn(n_neighbors=4, p=2, metric='minkowski')
classifier.fit(X_train,y_train)

y_pred=classifier.predict(X_test)
print(y_pred)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))



from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))

drugtree=DecisionTreeClassifier(criterion="entropy",max_depth=4)
drugtree
drugtree.fit(X_train,y_train)
pred=drugtree.predict(X_test)
print(pred[0:10])
print(y_test[0:10])

from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, pred))

from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)

yhat = clf.predict(X_test)
print(yhat)

from sklearn.metrics import f1_score
print(f1_score(y_test, yhat, average='weighted'))
```
